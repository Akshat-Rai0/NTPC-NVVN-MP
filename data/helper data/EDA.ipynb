```python
temp_cols = [
    "apparent_temperature_bhopal",
    "apparent_temperature_gwalior",
    "apparent_temperature_indore",
    "apparent_temperature_jabalpur",
    "apparent_temperature_singrauli",
    "apparent_temperature_ujjain"
]

df = df.drop(columns=temp_cols)
```

```python
print(df.columns)
```

```python
df = df.drop(columns=["is_holiday"])
```

```python
df.head()
```

```python
df_hourly = df.copy()
```

```python
#creation of date time index
df_hourly["datetime"] = pd.to_datetime(
    df_hourly["date"].astype(str)
    + " "
    + df_hourly["hour"].astype(str)
    + ":00:00"
)

df_hourly = df_hourly.set_index("datetime")
```

```python
df_15min = df_hourly.resample("15min").interpolate(
    method="linear"
)
```

```python
print(df_hourly.shape)

print(df_15min.shape)
```

```python
df_hourly.to_csv(
    "mp_hourly_cleaned.csv"
)

df_15min.to_csv(
    "mp_15min_interpolated.csv"
)
```

```python
df_15min.head(25)
```

```python
BASE_TEMP = 18

df_15min["CDD"] = (
    df_15min["temp_weighted"] - BASE_TEMP
).clip(lower=0)

df_15min["HDD"] = (
    BASE_TEMP - df_15min["temp_weighted"]
).clip(lower=0)
```

```python
df_15min["hour"] = (
    df_15min.index.hour +
    df_15min.index.minute/60
)
```

```python
df_15min.head()
```

```python
df_15min["season"] = df_15min["season"].ffill()
df_15min["holiday"] = df_15min["holiday"].ffill()
df_15min["month"] = df_15min["month"].ffill()
```
